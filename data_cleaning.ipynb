```python
import pandas as pd
import numpy as np
import re
```

```python
col_names = ['ProductName', 'Processor', 'RAM', 'OS', 'Disk', 'Dim']
df = pd.read_csv("data/ebay.csv", names=col_names, index_col = False)
```

```python
df.head()
```

### Get Product Brand

```python
# Clean column ProductName

ProductBrand =df['ProductName'].str.split(' ',1).loc[:].map(lambda x: x[0])
```

### Get Product Version

```python

productversion = df['ProductName'].str.split(' ',1).loc[:].map(lambda x: x[1])
ProdVer = productversion.str.split('Core').loc[:].map(lambda x: x[0])
```

```python
df = df.drop(columns ='ProductName')
df.insert(0, "ProductBrand", ProductBrand, True)
df.insert(1, "ProductVer", ProdVer, True)
df.head()
```

### Standardize Product Version

```python
#load standard version format for product
standard_versions = pd.read_csv('data/laptop_version.csv')
standard_versions.head()
```

```python
df['Flag'] = False  # inserting this column to keep track of product version standardization
```

```python
df.head()
```

```python
df.ProductVer = df.ProductVer.str.lower()
for i in range(len(standard_versions)):
    brand = standard_versions.Brand[i]
    version = standard_versions.Model[i]
    #print(str(version),str(brand))
    #df.loc[(df.ProductBrand).str.contains]
    df.loc[(df.ProductBrand.str.contains(str(brand),flags=re.IGNORECASE)) & (df.ProductVer.str.contains(str(version),flags=re.IGNORECASE)), 'ProductVer'] = str(version)
    df.loc[(df.ProductBrand.str.contains(str(brand),flags=re.IGNORECASE)) & (df.ProductVer.str.contains(str(version),flags=re.IGNORECASE)), 'Flag'] = True
```

```python
df[df['Flag'] == False].count()
```

```python
df[df['Flag'] == True].count()
```

```python
df[df['Flag'] == False]
```

### Assigning standard model to the laptops with Flag False

```python
model = []
AllBrands = df.loc[df['Flag']== False].ProductBrand
for i in AllBrands:
    model.append(standard_versions.Model[standard_versions.Brand == i].sample(n=1).to_string(index = False))
    
```

```python
count = 0
for i in df.loc[df['Flag'] == False].index:
    df.loc[i, 'ProductVer'] = model[count]
    df.at[i, 'ProductVer']  = model[count]
    count += 1
```

```python
df[df['Flag'] == False]
```

```python
# removing RAM, Display , Operating System like words from the description
df['RAM'] = df['RAM'].map(lambda x: x.rstrip('RAM'))
df['OS'] = df['OS'].map(lambda x: x.rstrip('Operating System'))
df['Dim'] = df['Dim'].map(lambda x: x.rstrip('Display'))
df.head()
```

```python
df = df.drop(columns = 'Flag')
```

```python
df.to_csv("data/ebay_clean.csv")
```
